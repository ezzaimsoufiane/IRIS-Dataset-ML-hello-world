# IMPORTS

```python
from keras.models import Sequential
from keras.layers import Dense
from sklearn import datasets
import numpy as np
from keras.utils import to_categorical
import random
from sklearn.model_selection import train_test_split

```

# The importing and formating

```python

iris=datasets.load_iris()
X= iris.data[:,[0,2]]
Y= iris.target
nY = to_categorical(Y)

print(X.shape)
print(Y.shape)
```

# The Magic


```python
random.seed(0)
X_train, X_test, Y_train, Y_test = train_test_split(X, nY, test_size=0.2, random_state=100)

model = Sequential()
model.add(Dense(128, input_dim=2, activation="tanh"))
model.add(Dense(64, activation="tanh"))
model.add(Dense(3, activation="softmax"))
model.compile(loss='categorical_crossentropy', optimizer="sgd", metrics=['accuracy', 'mse'])
model.fit(X_train,Y_train,validation_data=(X_test,Y_test),epochs=1500,batch_size=10)
```
